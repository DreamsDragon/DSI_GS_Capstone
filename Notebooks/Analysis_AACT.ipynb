```python
def sum_adverse_effect(row):
    try:
        return (sum([int(i) for i in str(row).split('|')]))
    except:
        return 0
```

```python
clinical_data['SumOtherEventStatsNumAffected'] = clinical_data['OtherEventStatsNumAffected'].apply(sum_adverse_effect)
clinical_data['SumOtherEventStatsNumAtRisk'] = clinical_data['OtherEventStatsNumAtRisk'].apply(sum_adverse_effect)
clinical_data['SumOtherEventStatsNumEvents'] = clinical_data['OtherEventStatsNumEvents'].apply(sum_adverse_effect)
clinical_data['SumSeriousEventStatsNumAffected'] = clinical_data['SeriousEventStatsNumAffected'].apply(sum_adverse_effect)
clinical_data['SumSeriousEventStatsNumAtRisk'] = clinical_data['SeriousEventStatsNumAtRisk'].apply(sum_adverse_effect)
clinical_data['SumSeriousEventStatsNumEvents'] = clinical_data['SeriousEventStatsNumEvents'].apply(sum_adverse_effect)
```

```python
clinical_data.head()
```

```python
clinical_data = clinical_data.drop(['OtherEventStatsNumAffected', 'OtherEventStatsNumAtRisk','OtherEventStatsNumEvents','SeriousEventStatsNumAffected','SeriousEventStatsNumAtRisk','SeriousEventStatsNumEvents'], axis = 1)
```

```python
clinical_data.drop(clinical_data.columns[[0, 10]], axis = 1, inplace = True)
```

```python
clinical_data
```

```python
clinical_data = clinical_data.groupby(['drug name','company name']).agg(
    {'SumOtherEventStatsNumAffected': 'sum',
     'SumOtherEventStatsNumAtRisk': 'sum',
     'SumOtherEventStatsNumEvents': 'sum',
     'SumSeriousEventStatsNumAffected': 'sum',
     'SumSeriousEventStatsNumAtRisk':'sum',
     'SumSeriousEventStatsNumEvents':'sum',
     'EnrollmentCount':'sum'})                                                                                                                                               
```

```python
clinical_data['clinical_rate'] = clinical_data['SumSeriousEventStatsNumAffected'] / clinical_data['EnrollmentCount']
```

```python
clinical_data = clinical_data[clinical_data['clinical_rate']  <= 1]  
```
