```python
import pandas as pd
import glob

path = r'C:\Users\User\Desktop\Columbia\Fall 21\Capstone\Data' # use your path
all_files = glob.glob(path + "/*.xlsx")
```

```python
li = []

for filename in all_files:
    df = pd.read_excel(filename)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame['year'] = pd.DatetimeIndex(frame['Approval Date']).year
frame['month'] = pd.DatetimeIndex(frame['Approval Date']).month
```

```python
frame_cl = frame.sort_values(by="Approval Date").drop_duplicates(subset=["Drug Name"], keep="last")
```

```python
frame_cl.to_excel("clean_approval_data.xlsx",sheet_name='clean_data')
frame.to_excel("raw_approval_data.xlsx",sheet_name='raw_data')
```
