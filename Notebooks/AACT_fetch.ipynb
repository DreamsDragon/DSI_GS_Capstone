```python
def sum_adverse_effect(row):
    try:
        return (sum([int(i) for i in str(row).split('|')]))
    except:
        return 0
```

# Clean approval data to fetch drug name and company name

```python
import regex as re
def clean_drug_name(df):
    try:
        drug_id = df['Drug Name'].split("#",1)[1]
        nda_split = df['Drug Name'].split("NDA",1)[0]
        if(' ' not in nda_split):
                drug_name = nda_split
                dosage = "None"
        elif('PLASTIC' in nda_split): 
                dosage = nda_split.split(' ',1)[1]
                drug_name = nda_split.split(' ',1)[0]
        else:
                drug_name = (re.split(r'(^[^\d]+)', nda_split)[1:])[0]
                dosage = (re.split(r'(^[^\d]+)', nda_split)[1:])[1]


        return(drug_name,dosage,drug_id)
    except:
        return
```

```python
def fetch_drug_clinical_data(drugname,companyname):
    url = "https://clinicaltrials.gov/api/query/study_fields?expr="+drugname+" "+companyname+"&fields=NCTId%2CBriefTitle%2CCondition%2COtherEventNotes%2CSeriousEventNotes%2COtherEventStatsNumAffected%2COtherEventStatsNumAtRisk%2COtherEventStatsNumEvents%2CSeriousEventStatsNumAffected%2CSeriousEventStatsNumAtRisk%2CSeriousEventStatsNumEvents%2CEnrollmentCount&min_rnk=1&max_rnk=1000&fmt=csv"
    r = requests.get(url)
    try:
        df = pd.read_csv(io.StringIO(r.text),skiprows = 10)
    except:
            df = pd.DataFrame()  
    if (df.empty):
        url = "https://clinicaltrials.gov/api/query/study_fields?expr="+drugname+"&fields=NCTId%2CBriefTitle%2CCondition%2COtherEventNotes%2CSeriousEventNotes%2COtherEventStatsNumAffected%2COtherEventStatsNumAtRisk%2COtherEventStatsNumEvents%2CSeriousEventStatsNumAffected%2CSeriousEventStatsNumAtRisk%2CSeriousEventStatsNumEvents%2CEnrollmentCount&min_rnk=1&max_rnk=1000&fmt=csv"
        r = requests.get(url)
        try:
            df = pd.read_csv(io.StringIO(r.text),skiprows = 10)
        except:
            print(drugname)
            df = pd.DataFrame()
            
    df["drug name"] = drugname
    df["company name"] = companyname
    return df
```

```python
master_clinical_data = []

for i in range(0,clean_data.shape[0]):
    df = fetch_drug_clinical_data(clean_data.loc[i]["Drug Name"], clean_data.loc[i]["Company"])
    master_clinical_data.append(df)
master_clinical_data = pd.concat(master_clinical_data)
        
    
 
```
